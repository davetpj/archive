## 1712

```python
A,B,C = map(int, input().split())

try:
    if B > C:
        print(-1)
    else:
        print(int(A/(C-B)+1))
except:
    print(-1)
```

## 2292

```python
%reset -f

N = int(input())

i,end_num,cnt = 1,1,1
while True:
    if N == 1:
        cnt = 0
        break

    end_num += i*6

    if N <= end_num:
        break
    cnt += 1
    i += 1

print(cnt+1)
```
