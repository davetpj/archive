## 1712

```python
A,B,C = map(int, input().split())

try:
    if B > C:
        print(-1)
    else:
        print(int(A/(C-B)+1))
except:
    print(-1)
```

## 2292

```python
%reset -f

N = int(input())

i,end_num,cnt = 1,1,1
while True:
    if N == 1:
        cnt = 0
        break

    end_num += i*6

    if N <= end_num:
        break
    cnt += 1
    i += 1

print(cnt+1)
```

## 1193

```python
%reset -f
# 같은 대각라인 행+열 값이 같음
# 출발지점 홀짝 구분
# 최종합에서 입력값을 뺀 후 인덱스 추출
X = int(input())
n,m = 0,0
while True:
    if m >= X:
        #print('인덱스:', X-(m-n))
        #print('층수:', n+1)
        if (n+1)%2 == 0:
            print(f'{((n+1) - (X-(m-n)))}/{X-(m-n)}')
        else:
            print(f'{X-(m-n)}/{((n+1) - (X-(m-n)))}')
        break
    n += 1
    m += n
```
