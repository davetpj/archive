## 2292

```python
%reset -f

N = int(input())

i,end_num,cnt = 1,1,1
while True:
    if N == 1:
        cnt = 0
        break

    end_num += i*6

    if N <= end_num:
        break
    cnt += 1
    i += 1

print(cnt+1)
```

## 1193

```python
%reset -f
# 같은 대각라인 행+열 값이 같음
# 출발지점 홀짝 구분
# 최종합에서 입력값을 뺀 후 인덱스 추출
X = int(input())
n,m = 0,0
while True:
    if m >= X:
        #print('인덱스:', X-(m-n))
        #print('층수:', n+1)
        if (n+1)%2 == 0:
            print(f'{((n+1) - (X-(m-n)))}/{X-(m-n)}')
        else:
            print(f'{X-(m-n)}/{((n+1) - (X-(m-n)))}')
        break
    n += 1
    m += n
```

## 10250

```python
# H 호텔 층수
# W 각 층의 방 수
# N 몇 번째 손님 
%reset -f
import math
T = int(input())
for _ in range(T):
    H,W,N = map(int, input().split())
    if N - (H*int(N/H)) == 0:
        if math.ceil(N/H) <10:
            print((f'{H}0{math.ceil(N/H)}'))
        else:
            print((f'{H}{math.ceil(N/H)}'))
    else:
        if math.ceil(N/H) <10:
            print((f'{N - (H*int(N/H))}0{math.ceil(N/H)}'))
        else:
            print((f'{N - (H*int(N/H))}{math.ceil(N/H)}'))
```

```python
%reset -f
t = int(input())

for _ in range(t):  
    floor = int(input())  # 층
    num = int(input())  # 호
    f0 = [x for x in range(1, num+1)]  # 0층 리스트
    for k in range(floor):  # 층 수 만큼 반복
        for i in range(1, num):  # 1 ~ n-1까지 (인덱스로 사용)
            f0[i] += f0[i-1]  # 층별 각 호실의 사람 수를 변경
    print(f0[-1])  # 가장 마지막 수 출력
```
