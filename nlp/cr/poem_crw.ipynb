```python
import requests
import pandas as pd
from tqdm import tqdm
from bs4 import BeautifulSoup
```

```python
#URL = "http://www.poemlove.co.kr/bbs/board.php?bo_table=tb22"

poet_list = ['김남식','김노연','김희달','박덕용','박성일','박인섭','유경란','유영호','유토비','유효주','이종화','이철용','임인규','임형섭','전금주','정연복','정윤칠','조동천','차영섭','최길준','최성찬','황인숙']

page_num = 1
headers = {"user-agent": "Mozilla/5.0"}

sub_url_list =[]

for poet in tqdm(poet_list):
    print(poet)
    sub_urls = []
    while True:
        print(page_num)
        URL = f"http://www.poemlove.co.kr/bbs/board.php?bo_table=tb22&sca={poet}&page={page_num}"

        url = URL
        res = requests.get(url, headers=headers)
        soup = BeautifulSoup(res.text, 'html.parser')
        body = soup.select_one('.list-board')
        
        if body.select_one('.wr-none'):
            break
        items = body.select(".wr-subject")
        for item in items[4:]:
            a=item.find("a")['href']
            #print(a)
            sub_urls.append(a)
        page_num += 1

    print(len(sub_urls))
    sub_url_list.append(sub_urls)
    page_num = 1

print(len(sub_url_list))
```

```python
titles,poets,contents = [],[],[]

for sub_urls in tqdm(sub_url_list):
    for sub_url in sub_urls:
        res = requests.get(sub_url, headers=headers)
        soup = BeautifulSoup(res.text, 'html.parser')
        article = soup.find('div',{'class':'view-wrap'})

        poet = article.find('span',{'class':'hidden-xs'}).text.strip()
        title = article.find('h1').text.strip()
        content = article.find('div',{'class':'view-content'}).text.strip()

        titles.append(title)
        poets.append(poet)
        contents.append(content)

print(len(titles),len(poets),len(contents))
```
